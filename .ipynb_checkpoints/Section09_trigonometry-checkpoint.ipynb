# VIDEO: Introduction to random numbers

```python
# Uniform distribution
nums = np.random.rand(1000)
plt.plot(nums, 's')
plt.show()

plt.hist(nums)
plt.show()
```

```python
# Uniform distribution -> values greater than 1
nums = np.random.rand(1000)
nums += 3
plt.plot(nums, 's')
plt.show()

plt.hist(nums)
plt.show()
```

```python
# Uniform distribution -> formula to have random value
nums = np.random.rand(1000)

min_val = 2
max_val = 17

nums = nums*(max_val - min_val) + min_val

plt.plot(nums, 's')
plt.show()

plt.hist(nums)
plt.show()
```

```python
# Normal distribution
nums = np.random.randn(1000)
plt.plot(nums, 's', alpha=.5)
plt.show()

plt.hist(nums)
plt.show()
```

### Exercise

```python
# generate normally distributed random numbers with a mean of 15 and a standard deviation of 4.3
desired_mean = 15
desired_std = 4.3

nums = desired_mean + desired_std*np.random.randn(1000)

plt.subplot(121)
plt.plot(nums,'s')
plt.xlabel('Index')
plt.ylabel('Value')

plt.subplot(122)
plt.hist(nums)
plt.xlabel('Value')
plt.ylabel('Bin count')
plt.show()

print('This distribution has a mean of %g and a standard deviation of %g' %(np.mean(nums),np.std(nums)))
```

```python
# to be more exact:

# generate normal random numbers
nums = np.random.randn(1000)

# enforce mean of 0
nums = nums - np.mean(nums)

# enforce std of 1
nums = nums/np.std(nums)

nums = nums*desired_std + desired_mean



plt.subplot(121)
plt.plot(nums,'s')
plt.xlabel('Index')
plt.ylabel('Value')

plt.subplot(122)
plt.hist(nums)
plt.xlabel('Value')
plt.ylabel('Bin count')
plt.show()

print('This distribution has a mean of %g and a standard deviation of %g' %(np.mean(nums),np.std(nums)))
```

# VIDEO: Exercise: Plotting random phase angles

```python

angs = np.random.randn(100) # not this one!
angs = np.random.rand(100)*2*np.pi

for i in angs:
    plt.plot([0,np.cos(i)],[0,np.sin(i)])

plt.axis('square')
plt.axis([-1,1,-1,1])
plt.axis('off')
plt.show()
```
