### Exercise
$ y = x^2 - 3x $ <br>
$ x \in \{-10, -9, ..., 10, 11\} $

```python
import matplotlib.pyplot as plt
```

```python
x = range(-10, 12)
for i in x:
    y = i**2 - 3*i
    plt.plot(i, y, 'o')

ylim = axis.get_ylim()
print(ylim)
axis.set_ylim([ylim[0], 120])
plt.show()
```

```python
import sympy as sym
```

```python
x = sym.symbols('x')
y = x**2 - 3*x

for pnti in range(-11, 12):
    plt.plot(pnti, y.subs(x, pnti), 'o')

plt.xlabel('x')
plt.ylabel('f(x) = %s' %y)
plt.show()
```

# VIDEO: Linear equations in slope-intercept form

```python
import matplotlib.pyplot as plt
```

```python
# y = mx + b
x = [-5, 5]
m = -8
b = 9

y = [432, 98439]

for i in range(0, len(x)):
    y[i] = m*x[i] + b
print(x, y)

plt.plot(x, y, label='y = %sx + %s' %(m, b))
plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
print(axis.get_xlim())
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.legend()

plt.show()
```

```python
# y = mx + b
# Using Numpy
import numpy as np

x = [-5, 5]
m = -8
b = 9

y = m*np.array(x) + b


plt.plot(x, y, label='y = %sx + %s' %(m, b))
plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
print(axis.get_xlim())
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.legend()

plt.show()
```

### Exercise

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [-5, 5]
m1 = 0.7
b1 = -2
```

```python
m2 = -1.25
b2 = 0.75
```

```python
y1 = m1*np.array(x) + b1
```

```python
y2 = m2*np.array(x) + b2
```

```python
plt.plot(x, y1, label='y = %sx + %s' %(m1, b1))
plt.plot(x, y2, label='y = %sx + %s' %(m2, b2))
plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
print(axis.get_xlim())
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.legend()
plt.title('The plot.')

plt.show()
```

```python
# Instructor solution
x = [-5, 5]
m = [.7, -1.25]
b = [-2, .75]

for i in range(0, len(m)):
    y = m[i] * np.array(x) + b[i]
    plt.plot(x, y, label=' y = %sx + %s' %(m[i], b[i]))


plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
#print(axis.get_xlim())
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.legend()
plt.title('The plot.')

plt.show()
```
